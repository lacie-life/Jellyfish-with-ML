```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/sonrasa2k/hongtin_dudoangianha/master/Real%20estate.csv")
```

```python
data.describe()
```

```python
## Plot data

import matplotlib.pyplot as plt
x = data["No"]
y = data["Y house price of unit area"]
plt.plot(x, y, 'go')
```

```python
names = ['X1 transaction date','X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude','X6 longitude']
X_train = data[names]
X_train
Y_train= data['Y house price of unit area']
Y_train
```

```python
!pip install sklearn
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, Y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
x_test = X_train
y_predict = reg.predict(x_test)
y_predict
```

```python
plt.scatter(x, y, color='black')
plt.plot(y_predict, color='blue', linewidth=2)
plt.xticks(())
plt.yticks(())
plt.show()
```
